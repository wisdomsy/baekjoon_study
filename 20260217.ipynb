```python
#11022
n=int(input())
A=[]
B=[]
for i in range(n):
  a,b=map(int,input().split())
  A.append(a)
  B.append(b)
for i in range(n):
  print('Case #',i+1,': ',A[i],' + ',B[i],' = ',A[i]+B[i],sep='')
```
