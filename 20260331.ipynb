```python
#1436 영화감독 숌

# 세자리 -> 1
# 네자리 -> 3개:5*2 / 4개:1
# 다섯자리 -> 3개:(5**2)*3 / 4개:5*2 / 5개:1
# 여섯자리 -> 3개:(5**3)*4 / 4개:(5**2)*3 / 5개:5*2 / 6개:1
# n자리 -> 3개:(5**(n-3))*(n-2) / 4개:(5**(n-4))*(n-3) / ... / n개:1

# 브루스포스
n=int(input())
number=666
count=0
while True:
  if '666' in str(number):
    count+=1
  if n==count:
    print(number)
    break
  number+=1
```
