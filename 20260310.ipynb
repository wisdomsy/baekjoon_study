```python
# 2581 소수
def prime(n):
  check=True
  if n==1:
    check=False
  for i in range(2,n):
    if n%i==0:
      check=False
      break
  if check==True:
    return True
  else:
    return False

m=int(input())
n=int(input())
sum=0
prime_list=[]
for i in range(m,n+1):
  if prime(i)==True:
    prime_list.append(i)
for i in range(len(prime_list)):
  sum+=prime_list[i]
if len(prime_list)==0:
  print(-1)
else:
  print(sum)
  print(prime_list[0])
```
