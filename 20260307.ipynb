```python
# 5086 배수와 약수
while True:
  a,b=map(int,input().split())
  if a==b==0:
    break
  if b%a==0:
    print('factor')
  elif a%b==0:
    print('multiple')
  else:
    print('neither')
```

```python
# 2501 약수 구하기
n,k=map(int,input().split())
count=0
for i in range(1,n+1):
  if n%i==0:
    count+=1
  if count==k:
    print(i)
    break
if count<k:
  print(0)
```
