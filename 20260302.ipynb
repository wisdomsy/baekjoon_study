```python
#2745 진법 변환
N,B=map(str,input().split())
sum=0
length=len(N)
for i in range(length):
  if ord(N[i])>=65:
    sum+=(int(ord(N[i]))-55)*(int(B)**(length-i-1))
  else:
    sum+=int(N[i])*(int(B)**(length-i-1))
print(sum)
```

```python
#11005 진법 변환 2
N,B=map(int,input().split())
result=''
while True:
  if N==0:
    break
  change=N%B
  N=N//B
  if change>=10:
    result+=chr(change+55)
  else:
    result+=str(change)
for i in range(len(result)-1,-1,-1):
  print(result[i],end='')
```
