```python
# 9506 약수들의 합
while True:
  small=[]
  n=int(input())
  if n==-1:
    break
  for i in range(1,n):
    if n%i==0:
      small.append(i)
  sum=0
  for i in range(len(small)):
    sum+=small[i]
  if n==sum:
    print(n,'=',end=' ')
    for i in range(len(small)-1):
      print(small[i],end=' ')
      print('+',end=' ')
    print(small[-1])
  else:
    print(n,'is NOT perfect.')
```

```python
# 1978 소수 찾기
n=int(input())
num=input()
num_list=num.split(' ')
length=len(num_list)
count=0
for i in range(length):
  count+=1
  if int(num_list[i])==1:
    count-=1
  else:
    for j in range(2,int(num_list[i])):
      if int(num_list[i])%j==0:
        count-=1
        break
print(count)
```
