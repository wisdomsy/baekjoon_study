```python
#10810
n,m=map(int,input().split())
result=[0 for i in range(n)]
for _ in range(m):
  i,j,k=map(int,input().split())
  for s in range(i-1,j):
    result[s]=k
for i in range(len(result)):
  print(result[i],end=' ')
```

```python
#10813
n,m=map(int,input().split())
result=[]
for i in range(n):
  result.append(i+1)
for _ in range(m):
  i,j=map(int,input().split())
  result[i-1],result[j-1]=result[j-1],result[i-1]
for i in range(len(result)):
  print(result[i],end=' ')
```

```python
#5597
result=[]
for i in range(28):
  result.append(int(input()))
result.sort()
miss=[]
for i in range(27):
  if result[i+1]-result[i]!=1:
    miss.append(result[i]+1)
if result[0]!=1:
  miss.append(1)
if result[-1]!=30:
  miss.append(30)
miss.sort()
for i in range(2):
  print(miss[i])
```
