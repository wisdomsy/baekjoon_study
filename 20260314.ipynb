```python
# 10101 삼각형 외우기

a=int(input())
b=int(input())
c=int(input())
if a==b==c==60:
  print('Equilateral')
elif a+b+c==180:
  if a==b or b==c or c==a:
    print('Isosceles')
  else:
    print('Scalene')
else:
  print('Error')
```

```python
# 5073 삼각형과 세 변
while True:
  a,b,c=map(int,input().split())
  if a==b==c==0:
    break
  if a==b==c:
    print('Equilateral')
  elif a>=b+c or b>=a+c or c>=a+b:
    print('Invalid')
  elif a==b or b==c or c==a:
    print('Isosceles')
  else:
    print('Scalene')
```
