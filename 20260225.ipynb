```python
#1152
s=input()
count=0
start=0
string=False
# 주어진 문자가 공백 자체인 경우 고려
for i in range(len(s)):
  if s[i]!=' ':
    string=True
if string==False:
  print(0)
else:
  for i in range(len(s)):
    if s[i]!=' ':
      start=i
      break
  for i in range(len(s)):
    if s[i]!=' ':
      end=i
  for i in range(start,end+1):
    if s[i]==' ':
      count+=1
  print(count+1)
```
