```python
# 3009 네 번째 점
a_list=[]
b_list=[]
for i in range(3):
  a,b=map(int,input().split())
  a_list.append(a)
  b_list.append(b)
  a_list.sort()
  b_list.sort()

if a_list[1]==a_list[0]:
  print(a_list[-1],end=' ')
else:
  print(a_list[0],end=' ')

if b_list[1]==b_list[0]:
  print(b_list[-1],end='')
else:
  print(b_list[0],end='')
```
