```python
# 2720 세탁소 사장 동혁
n=int(input())
for i in range(n):
  money=int(input())
  result=[]
  # float형으로 처리하면 부동소수점 문제 생길 수 있음
  # int형으로 처리하는 게 안전함
  result.append(money//25)
  money-=(money//25)*25
  result.append(money//10)
  money-=(money//10)*10
  money=round(money,2)
  result.append(money//5)
  money-=(money//5)*5
  money=round(money,2)
  result.append(money//1)
  money-=(money//1)*1
  for i in range(4):
    print(int(result[i]),end=' ')
```
