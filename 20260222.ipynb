```python
#10811
n,m=map(int,input().split())
basket=[]
for i in range(1,n+1):
  basket.append(i)

# 인덱스가 꼬여서 오류
# for _ in range(m):
#   start,end=map(int,input().split())
#   if start!=end:
#     for i in range(start-1,(end+start)//2):
#       for j in range((end-start)//2+1):
#         basket[i+j],basket[end-1-j]=basket[end-1-j],basket[i+j]

# 인덱스 자체를 새로 정의하여 재정의
for _ in range(m):
    start, end = map(int, input().split())
    i = start - 1
    j = end - 1
    while i < j:
        basket[i], basket[j] = basket[j], basket[i]
        i += 1
        j -= 1
for i in range(n):
  print(basket[i],end=' ')
```
