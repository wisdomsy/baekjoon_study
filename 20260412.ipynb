```python
# 11650 좌표 정렬하기
n=int(input())
x_list=[]
y_list=[]
for i in range(n):
  x,y=map(int,input().split())
  x_list.append(x)
  y_list.append(y)

for i in range(n-1):
  if x_list[i]>x_list[i+1]:
    x_list[i],x_list[i+1]=x_list[i+1],x_list[i]
    y_list[i],y_list[i+1]=y_list[i+1],y_list[i]
  if x_list[i]==x_list[i+1]:
    if y_list[i]>y_list[i+1]:
      y_list[i],y_list[i+1]=y_list[i+1],y_list[i]

for i in range(n):
  print(x_list[i],y_list[i])
```
