```python
n, m = map(int, input().split())

board = []
for i in range(n):
    board.append(input())

results = []

# 8x8 결정
for i in range(n - 7):
    for j in range(m - 7):
        draw1 = 0 # 'W'로 시작하는 경우
        draw2 = 0 # 'B'로 시작하는 경우

        # 8x8 영역만 검사
        for a in range(i, i + 8):
            for b in range(j, j + 8):
                if (a + b) % 2 == 0:
                    if board[a][b] != 'W':
                        draw1 += 1
                    if board[a][b] != 'B':
                        draw2 += 1
                else:
                    if board[a][b] != 'B':
                        draw1 += 1
                    if board[a][b] != 'W':
                        draw2 += 1

        results.append(draw1)
        results.append(draw2)
print(min(results))
```
