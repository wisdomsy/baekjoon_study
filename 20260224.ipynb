```python
#10809
alphabet=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
result=[-1 for i in range(26)]
s=input()
for i in range(len(alphabet)):
  for j in range(len(s)):
    if s[j]==alphabet[i]:
      result[i]=j
      break
for i in range(26):
  print(result[i],end=' ')
```
